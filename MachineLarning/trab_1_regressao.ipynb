## Fase 1
Implemente duas funções chamadas correlacao e regressao. Cada uma deve ter dois vetores Nx1 como entrada, onde N é a dimensão do vetor (no caso de x N=11). A primeira função deve calcular o coeficiente de correlação r, e a segunda função deve calcular a regressão, isto é, β0 e β1.

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def plotDados(b0, b1, xvetor, yvetor, correlacao):
  regressao = []
  for x in xvetor:
    regressao.append(b0 + b1 * x)

  plt.scatter(xvetor, yvetor)
  plt.plot(xvetor, regressao)
  plt.title(f"X vs. Y - onde correlação = {correlacao} ")
  plt.suptitle(f"b0 = {b0}, b1 = {b1}")
  plt.xlabel("X")
  plt.ylabel("Y")

  plt.show()

def calculaMedia (vetor):
  media = sum(vetor) / float(len(vetor))
  return media

def soma(vetorx, vetory, mediax, mediay):
  soma = 0
  for x, y in zip(vetorx, vetory):
    soma += (x - mediax) * (y - mediay)

  return soma

def calculaCorrelacao(vetorx, vetory, xmedia, ymedia, soma1):
  somax = 0
  somay = 0
  for x, y in zip(vetorx, vetory):
    somax += (pow(x - xmedia, 2))
    somay += (pow(y - ymedia, 2))

  raiz = math.sqrt(somax * somay)
  correlacao = soma1 / raiz

  return float(correlacao), somax

# Datafases providos pela prof
x1 = [10,8,13,9,11,14,6,4,12,7,5];
y1 = [8.04,6.95,7.58,8.81,8.33,9.96,7.24,4.26,10.84,4.82,5.68];


x2 = [10,8,13,9,11,14,6,4,12,7,5];
y2 = [9.14,8.14,8.47,8.77,9.26,8.10,6.13,3.10,9.13,7.26,4.74];


x3 = [8,8,8,8,8,8,8,8,8,8,19];
y3 = [6.58,5.76,7.71,8.84,8.47,7.04,5.25,5.56,7.91,6.89,12.50];

# calculo correlação

# 1 - tirar média
x1Media = calculaMedia(x1)
y1Media = calculaMedia(y1)

x2Media = calculaMedia(x2)
y2Media = calculaMedia(y2)

x3Media = calculaMedia(x3)
y3Media = calculaMedia(y3)

# 2 - soma
soma1 = soma(x1, y1, x1Media, y1Media)
soma2 = soma(x2, y2, x2Media, y2Media)
soma3 = soma(x3, y3, x3Media, y3Media)

# 3 - faz a raiz quadrada e divide = resultado de correlação
correlacao1, somax1 = calculaCorrelacao(x1, y1, x1Media, y1Media, soma1)
correlacao2, somax2 = calculaCorrelacao(x2, y2, x2Media, y2Media, soma2)
correlacao3, somax3 = calculaCorrelacao(x3, y3, x3Media, y3Media, soma3)

# calculo regressão
b11 = soma1 / somax1
b01 = y1Media - b11 * x1Media

b12 = soma2 / somax2
b02 = y2Media - b12 * x2Media

b13 = soma3 / somax3
b03 = y3Media - b13 * x3Media

# mostra o gráfico
plotDados(b01, b11, x1, y1, correlacao1)
plotDados(b02, b12, x2, y2, correlacao2)
plotDados(b03, b13, x3, y3, correlacao3)
```
