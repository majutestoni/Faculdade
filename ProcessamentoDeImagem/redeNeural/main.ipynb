```python
# Maria Júlia Testoni
# Perceptron

import numpy as np
import pandas as pd

dados = {
    'Atleta': ['Neymar', 'Messi', 'Barichello', 'Massa'],
    'Ciência': [1, 1, 1, 1],
    'Física': [0, 0, 1, 1],
    'Filosofia': [0, 1, 0, 1],
    'Classe': [0, 0, 1, 1]
}

# Dados novos pra testar depois que o modelo treinar
dados_teste = {
    'Atleta': ['Pessoa 1', 'Pessoa 2', 'Pessoa 3', 'Pessoa 4'],
    'Ciência': [0, 0, 1, 1],
    'Física':  [0, 1, 0, 1],
    'Filosofia': [0, 1, 1, 0]
}

df = pd.DataFrame(dados)
# Separando as colunas que vão virar entrada (X) e a saída (y)
x = df[['Ciência', 'Física', 'Filosofia']].values
y = df['Classe'].values

df_teste = pd.DataFrame(dados_teste)
x_teste = df_teste[['Ciência', 'Física', 'Filosofia']].values

class Perceptron:
    def __init__(self, n_iter=10, learning_rate=1):
        self.n_iter = n_iter
        self.learning_rate = learning_rate
    
    def fit(self, x, y):
        self.w_ = np.zeros(x.shape[1] + 1)
        self.errors_ = []

        for epoch in range(self.n_iter):
            errors = 0
            for xi, target in zip(x, y):
                xi_bias = np.insert(xi, 0, 1) 
                
                # cálculo Σ = e1*peso1 + e2*peso2 + e3*peso3 
                soma = 0
                for entrada, peso in zip(xi_bias, self.w_):
                    soma += entrada * peso

                output = 1 if soma >= 0 else 0

                if output != target:
                    errors += 1
                    if target == 1:
                        # Se =1 && erro, soma os sinais de entrada * taxa de aprendizado
                        self.w_ += self.learning_rate * xi_bias
                    else:
                        # Se =0 && erro, subtrai os sinais de entrada * taxa de aprendizado
                        self.w_ -= self.learning_rate * xi_bias

            self.errors_.append(errors)

    def predict(self, X):
        X_bias = np.insert(X, 0, 1, axis=1)
        soma = np.dot(X_bias, self.w_)
        return np.where(soma >= 0, 1, 0)

perceptron = Perceptron(n_iter=10, learning_rate=1)
perceptron.fit(x, y)

df['Classe prevista'] = perceptron.predict(x)

print("Pesos finais:", perceptron.w_)
print("\nTabela após treinamento:\n")
print(df)


df_teste['Classe prevista'] = perceptron.predict(x_teste)
print("\nTabela de testes:\n")
print(df_teste)
```
